# Crackle Pop

Below I write a program in Python that prints out the numbers 1 to 100 (inclusive). If the number is divisible by 3, it prints Crackle instead of the number. If it's divisible by 5, it prints Pop. If it's divisible by both 3 and 5, it prints CracklePop. I submitted this prompt to the Recurse Center.

```python
for i in range(1, 101):
    if((i%3==0) & (i%5==0)):
        print('CracklePop')
    if (i%3==0):
        print('Crackle')
    if (i%5==0):
        print('Pop')
    else:
        print(i)
```
